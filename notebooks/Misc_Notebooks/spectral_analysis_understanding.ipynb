### Sine Waves

$a * sin(2\pi ft + \theta)$

Three parameters:
1. a: amplitude
2. f: frequency
3. $\theta$: phase

None of the parameters determine other parameters. So, they are independent of each other.

**Frequency** 

- Frequency refers to the number of complete cycles of a periodic event occurring per unit time.
- It is the inverse of the period.
- The period is the time taken between two consecutive peaks, two consecutive troughs, or two consecutive identical points in a wave.

**Phase**

- Value of the sine wave as it crosses the point corresponding to time equals zero.

**Amplitude**

Note: This is the not the instantaneous value of the signal at each point in time.

- The peak-to-peak amplitude is the total distance from the maximum positive peak to the maximum negative trough of the waveform.
- The amplitude of a wave, in this context, is defined as half of the peak-to-peak amplitude. It represents the maximum displacement from the equilibrium (zero) position to the peak.


```python
# simulation parameters
srate = 100; # sampling rate in Hz
time  = np.arange(0.,2.,1./srate) # time in seconds

# sine wave param.eters
freq = 3;    # frequency in Hz
ampl = 2;    # amplitude in a.u.
phas = np.pi/3; # phase in radians

# generate the sine wave
sinewave = ampl * np.sin( 2*np.pi * freq * time + phas )

plt.plot(time,sinewave,'k')
plt.xlabel('Time (sec.)')
plt.ylabel('Amplitude (a.u.)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq

# Time array for 2 seconds
t = np.linspace(0, 2, 2000)

# Simulate a simple sinusoidal wave to represent a 1 Hz heartbeat
heartbeat_frequency = 1  # 1 Hz
simple_heartbeat = np.sin(2 * np.pi * heartbeat_frequency * t)

# Simulate a simple ECG-like signal (note this is a very simplified version)
ecg_signal = np.sin(2 * np.pi * 1 * t) + 0.5 * np.sin(2 * np.pi * 50 * t) + 0.25 * np.sin(2 * np.pi * 75 * t)

# Perform FFT on the ECG-like signal
N = len(t)  # Number of samples
T = t[1] - t[0]  # Sample spacing
yf = fft(ecg_signal)
xf = fftfreq(N, T)[:N//2]

plt.figure(figsize=(12, 8))

# Plot simple sine wave
plt.subplot(3, 1, 1)
plt.plot(t, simple_heartbeat)
plt.title("Simple 1 Hz Sine Wave (Heartbeat)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot ECG-like signal in time domain
plt.subplot(3, 1, 2)
plt.plot(t, ecg_signal)
plt.title("Simplified ECG-like Signal (Time Domain)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot ECG-like signal in frequency domain
plt.subplot(3, 1, 3)
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.title("Simplified ECG-like Signal (Frequency Domain)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
